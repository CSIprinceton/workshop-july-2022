```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rdf1 = []
Nbin = 100
Nsamp = 1001
with open("./900K/vii.rdf", "r") as f:
    for i in range(3):
        f.readline()
    for i in range(Nsamp):
        f.readline()
        for j in range(Nbin):
            rdf1.append(f.readline().split())
rdf1 = np.array(rdf1, dtype = "float").reshape(Nsamp, Nbin, 8)
rdf1_mean = rdf1[:].mean(axis=0)
```

```python
rdf2 = []
Nbin = 100
Nsamp = 1001
with open("./1300K/vii.rdf", "r") as f:
    for i in range(3):
        f.readline()
    for i in range(Nsamp):
        f.readline()
        for j in range(Nbin):
            rdf2.append(f.readline().split())
rdf2 = np.array(rdf2, dtype = "float").reshape(Nsamp, Nbin, 8)
rdf2_mean = rdf2[:].mean(axis=0)
```

```python
rdf3 = []
Nbin = 100
Nsamp = 1001
with open("./1800K/vii.rdf", "r") as f:
    for i in range(3):
        f.readline()
    for i in range(Nsamp):
        f.readline()
        for j in range(Nbin):
            rdf3.append(f.readline().split())
rdf3 = np.array(rdf3, dtype = "float").reshape(Nsamp, Nbin, 8)
rdf3_mean = rdf3[:].mean(axis=0)
```

```python
plt.figure(figsize=(8, 3.4))
plt.xlim(0, 6)
plt.ylim(0, 4.0)
plt.xlabel("$r/\mathrm{\AA}$", fontsize=14)
plt.ylabel("$g_{\mathrm{OO}}(r)$", fontsize=14)
plt.xticks(np.arange(0, 7, 1))
plt.plot(rdf1_mean[:, 1], rdf1_mean[:, 2], color = "green", label="ice VII")
plt.plot(rdf2_mean[:, 1], rdf2_mean[:, 2], color = "blue", label="ice VII$^{\prime\prime}$")
plt.plot(rdf3_mean[:, 1], rdf3_mean[:, 2], color = "red", label="ionic fluid")

plt.legend(fontsize=14)
plt.plot([0, 6], [0.5, 0.5], linestyle = "--", color = "black")
plt.plot([0, 6], [1.0, 1.0], linestyle = "--", color = "black")
plt.plot([0, 6], [1.5, 1.5], linestyle = "--", color = "black")
plt.plot([0, 6], [2.0, 2.0], linestyle = "--", color = "black")
plt.plot([0, 6], [2.5, 2.5], linestyle = "--", color = "black")
plt.plot([0, 6], [3.0, 3.0], linestyle = "--", color = "black")
plt.plot([0, 6], [3.5, 3.5], linestyle = "--", color = "black")
```

```python
plt.figure(figsize=(8, 3.4))
plt.xlim(0, 6)
plt.ylim(0, 2.5)
plt.xlabel("$r/\mathrm{\AA}$", fontsize=14)
plt.ylabel("$g_{\mathrm{OH}}(r)$", fontsize=14)
plt.xticks(np.arange(0, 7, 1))
plt.yticks(np.arange(0, 2.5, 0.5))
plt.plot(rdf1_mean[:, 1], rdf1_mean[:, 4], color = "green", label="ice VII")
plt.plot(rdf2_mean[:, 1], rdf2_mean[:, 4], color = "blue", label="ice VII$^{\prime\prime}$")
plt.plot(rdf3_mean[:, 1], rdf3_mean[:, 4], color = "red", label="ionic fluid")

plt.legend(fontsize=14)
plt.plot([0, 6], [0.5, 0.5], linestyle = "--", color = "black")
plt.plot([0, 6], [1.0, 1.0], linestyle = "--", color = "black")
plt.plot([0, 6], [1.5, 1.5], linestyle = "--", color = "black")
plt.plot([0, 6], [2.0, 2.0], linestyle = "--", color = "black")
plt.plot([0, 6], [2.5, 2.5], linestyle = "--", color = "black")
```
