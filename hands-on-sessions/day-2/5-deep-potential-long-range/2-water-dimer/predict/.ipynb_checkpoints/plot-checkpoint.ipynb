```python
import matplotlib.pyplot as plt
import numpy as np

d,dp,dplr,dft = [],[],[],[]

with open('dplr/dplr.data.out', 'r') as f1:
    lines = f1.readline()[1:]
    lines = f1.readlines()
    for line in lines:
        value = [float(s) for s in line.split()]
        d.append(value[0])
        dplr.append(value[1])
        dft.append(value[2])

with open('dp/dp.data.out', 'r') as f2:
    lines = f2.readline()[1:]
    lines = f2.readlines()
    for line in lines:
        value = [float(s) for s in line.split()]
        dp.append(value[1])


        
plt.figure(figsize=(8, 3.4))
plt.xlim(2, 10)
plt.ylim(-0.25, 0.05)
plt.xlabel("d (angstrom)", fontsize=14)
plt.ylabel("E (eV)", fontsize=14)
plt.plot(d, dft, color = "black", label="DFT")
plt.plot(d, dp, color = "red", label="DP")
plt.plot(d, dplr, color = "blue", label="DPLR")
plt.legend(fontsize=14)
```
